```python
import pandas as pd
import yfinance as yf
```

```python
ohlcv_data={}
```

```python
tickers={"AMZN","GOOG","MSFT"}
```

```python
for ticker in tickers:
    temp=yf.download(ticker,period="1mo",interval="5m")
    temp.dropna(how="any",inplace=True)
    ohlcv_data[ticker]=temp
```

```python
def Bollinger_band(DF,n=14):
    df=DF.copy()
    df["MB"]=df["Adj Close"].rolling(n).mean()
    df["UB"]=df["MB"]+2*df["Adj Close"].rolling(n).std(ddof=0) #degree of freedom = 0 
    df["LB"]=df["MB"]-2*df["Adj Close"].rolling(n).std(ddof=0)
    df["Band_width"]=df["UB"] - df["LB"]
    return df[["MB","UB","LB","Band_width"]]
```

```python
for ticker in ohlcv_data:
    ohlcv_data[ticker][["MB","UB","LB","Band_width"]] = Bollinger_band(ohlcv_data[ticker])
```
