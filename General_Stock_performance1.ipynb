```python
import datetime as dt
import yfinance as yf
import pandas as pd
```

```python
tickers=["AMZN","MSFT","FB","GOOG"]
```

```python
start=dt.datetime.today()-dt.timedelta(3650)
end=dt.datetime.today()
```

```python
cl_price=pd.DataFrame()
```

```python
for ticker in tickers:
    cl_price[ticker]=yf.download(ticker,start,end)["Adj Close"]
```

```python
cl_price.dropna(axis=0,how="any",inplace=True)
```

```python
daily_return=cl_price.pct_change()
```

```python
cl_price.plot(subplots=True,layout=(2,2),title="Stock Price Evolution",grid=True)
```

```python
daily_return.plot(subplots=True,layout=(2,2),title="Daily Return Evolution",grid=True)
```
